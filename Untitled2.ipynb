```python
import pandas as pd
import ipywidgets as widgets
from IPython.display import display, clear_output

# Leer CSV del repositorio
df = pd.read_csv("download (1).csv")

# Widgets de entrada y salida
busqueda = widgets.Text(
    placeholder='Cliente o Patente...',
    description='Buscar:',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='70%')
)

boton_buscar = widgets.Button(description="üîç Buscar", button_style='primary')
salida = widgets.Output()

# Funci√≥n que se ejecuta al hacer clic
def buscar(b):
    with salida:
        clear_output()
        termino = busqueda.value.strip().lower()
        if not termino:
            print("‚ö†Ô∏è Ingres√° un t√©rmino para buscar.")
            return
        resultados = df[df.apply(lambda row:
            termino in str(row['Patente']).lower() or
            termino in str(row['Cliente']).lower(), axis=1)]
        if resultados.empty:
            print("‚ùå No se encontraron coincidencias.")
        else:
            display(resultados)

# Conexi√≥n del bot√≥n con la funci√≥n
boton_buscar.on_click(buscar)

# Mostrar interfaz
display(widgets.VBox([busqueda, boton_buscar, salida]))
```
